```python
%matplotlib inline

from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import yaml
import os
import zipfile
import gdown
```

```python
file_id = "1ItS0f_v8G4iSIB-PvRrxVwGcCWZRkFuC"
url = f"https://drive.google.com/uc?id={file_id}"

gdown.download(url, "yolo_seg_dataset_8.zip", quiet=False)
```

```python
with zipfile.ZipFile("yolo_seg_dataset_8.zip", 'r') as zip_ref:
    zip_ref.extractall("data")
```

```python
data_yaml = {
    'path': '/kaggle/working/data/yolo_seg_dataset_8',
    'train': 'images/train',
    'val': 'images/val',
    'names': {
          0: "laptop",
          1: "mouse",
          2: "keyboard",
          3: "mobile",
          4: "monitor" 
    }
}

with open('data.yaml', 'w') as f:
    yaml.dump(data_yaml, f)

print("data.yaml created ✅")
```

```python
model = YOLO("yolov8s-seg.pt")
```

```python
# imgsz = 832 or 896
```

```python
model.train(
    data="data.yaml",
    epochs=100,
    imgsz=832, #should for exp2
    batch=16,
    device=0,
    workers=2,
    cache=True,
    half=True,
    cos_lr=True,
    patience=20,
    mosaic=0.5,
    mixup=0.0,
    project="runs/segment",
    name="exp1",         # experiment name
    exist_ok=True        # overwrite if exists
)
```
